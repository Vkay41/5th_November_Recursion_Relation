1. Assume a solution of the form T(n) = A . n + B
2. Substitute the assumed solution into the recurrence relation.
3. Solve for the constants A and B.

T(n) = T(n-1) + c

A . n + B = A . n - A + B + c 
A = c

The solution is T(n) = c . n + B
